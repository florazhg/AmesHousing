### remove unecessary columns

```python
df.drop('Order',axis=1,inplace=True)
df.drop('PID',axis=1,inplace=True)
```

## 2) Dealing with missing values

```python
# drop columns with more than 60% missing values
col_to_drop = nan_percentage[nan_percentage>60].keys() # list of columns to drop
df.drop(col_to_drop, axis=1, inplace=True)
```

```python
# check again for missing values
nan_df = df.isna().sum()
nan_df[nan_df>0]
```

```python
# keep the nan in a list
nan_columns = nan_df[nan_df>0].keys().tolist()
nan_columns
```

```python
# check the type of the nan columns
df[nan_columns].dtypes
```

```python
# for the categorical columns, fill the missing values with 'missing'
# for the numerical columns, fill the missing values with the median
for col in nan_columns :
    if df[col].dtypes == 'object' :
        df[col] = df[col].fillna('Missing')
    else :
        median_col = df[col].median()
        df[col] = df[col].fillna(median_col)
```

```python
df.isna().sum()
```

### Convert type

```python
df['Garage Yr Blt']=df['Garage Yr Blt'].astype(int)
```

```python
df['Garage Cars']=df['Garage Cars'].astype(int)
```

### Convert date-time values

```python
# identify the year columns
df.columns
```

```python
# keep the year columns in a list
year_columns = df[['Year Built','Year Remod/Add','Garage Yr Blt','Yr Sold']].keys().tolist()
print(year_columns)
```

```python
# convert to year only
for col in year_columns:
    df[col] = pd.to_datetime(df[col], format="%Y")
    df[col]=df[col].dt.year
df.info()
```
